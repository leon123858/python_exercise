# 利用 mean 法分成k群

```python
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
#從sklearn的資料庫裡面下載鳶尾花資料集，僅使用特徵的部分，為了方便視覺化，我們今天##僅採用鳶尾花花瓣的長度及寬度兩個特徵
iris=datasets.load_iris()
X=iris.data
X=X[:,2:4]
```

```python
#模型建立與分群
KM=KMeans(n_clusters=3,init='random',random_state=5)
KM.fit(X)
KM.predict(X)
```

```python
#圖像化顯示結果
plt.scatter(X[:,0],X[:,1],c=KM.predict(X))
```

# K-Means with random data

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
#建立資料，選擇樣本數、變數、群組、群組標準差、隨機種子
data = make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.8,random_state=101)
```

```python
#在data[0]中，可以看到這是一個二維陣列
#[:,0]=[第一橫排:最後橫排 ,第0行（或稱第一個變數）]
#[:,1]=[第一橫排:最後橫排 ,第1行（或稱第二個變數）]
#data[1]代表分群結果
#print(data[0])
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

```python
#設定分成四組（因為知道原始資料是分成四組，但改成其他分組數字，就可以從下方圖表比較出差異）
kmeans = KMeans(n_clusters=4)

#使用KMeans演算法來建立模型
kmeans.fit(data[0])

#因為分成四組，所以centroid的點也有四個
kmeans.cluster_centers_

#分群結果
kmeans.labels_
```

```python
#plt.subplot(橫排數量,每個橫排有幾個圖並列)
#sharey=True共享y軸
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))
ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')
ax2.set_title("Original")
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

```python
#分群成5 比較
#設定分成5組
kmeans2 = KMeans(n_clusters=5)

#使用KMeans演算法來建立模型
kmeans2.fit(data[0])
#draw
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))
ax1.set_title('K Means 4')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')
ax2.set_title('K Means 5')
ax2.scatter(data[0][:,0],data[0][:,1],c=kmeans2.labels_,cmap='rainbow')
```

# DBSCAN 密度分群法

```python
#https://scikit-learn.org/stable/modules/generated/sklearn.cluster.DBSCAN.html
from sklearn.cluster import DBSCAN
import numpy as np
X = data[0]
clustering = DBSCAN(eps=1, min_samples=4).fit(X) #eps表示密度測量圈的半徑 , min_sample表示成立群的最小樣本數
print(clustering.labels_) #結果的-1表示雜訊 不被任何群包含
#print(clustering)
```

# Agglomerative  聚合式階層分群法
### sklearn.cluster.AgglomerativeClustering

```python
#https://scikit-learn.org/stable/modules/generated/sklearn.cluster.AgglomerativeClustering.html
from sklearn.cluster import AgglomerativeClustering
import numpy as np
X = data[0]
clustering = AgglomerativeClustering().fit(X)
#clustering
#AgglomerativeClustering()
print(clustering.labels_)
#array([1, 1, 1, 0, 0, 0])
```
