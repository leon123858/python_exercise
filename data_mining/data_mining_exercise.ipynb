```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.frequent_patterns import fpgrowth
```

```python
dataset = [['Milk', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
 ['Dill', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
 ['Milk', 'Apple', 'Kidney Beans', 'Eggs'],
 ['Milk', 'Unicorn', 'Corn', 'Kidney Beans', 'Yogurt'],
 ['Corn', 'Onion', 'Onion', 'Kidney Beans', 'Ice cream', 'Eggs']]
```

```python
#資料轉換
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
print(df)
```

# apriori 進行關聯分析

```python
#關聯 patern 和其 機率/條件機率
frequent_items = apriori(df,min_support = 0.6,use_colnames = True)
frequent_items
```

```python
#關聯分析 #lift 表示相關程度 >1 表示正相關  =1 表示獨立
rules = association_rules(frequent_items,metric="confidence",min_threshold=0.6)
rules
```

# FP Growth algorithm 進行關聯分析

```python
#fpgrowth找出來的關聯樣式和機率
table1 = fpgrowth(df, min_support=0.6, use_colnames=True)
table1
```

```python
#再找一次關聯性同apriori
rules = association_rules(table1,metric="confidence",min_threshold=0.6)
rules
```

```python
#設定自訂關聯規則
rules = association_rules(table1,metric="lift",min_threshold=1)
rules
```
